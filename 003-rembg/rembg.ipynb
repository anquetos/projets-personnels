# Rembg

Test de Rembg, un outil permettant de supprimer l'arrière plan d'une image.

```python
# Importe la librairie rembg
import rembg
from rembg import remove
print('rembg version', rembg.__version__)

# Importe la librairie Matplotlib
import matplotlib.pyplot as plt
```

```python
# Configure les chemins d'accès
input_path = './img/input.jpg'
output_path = './img/output.png'
```

```python
# Lecture et affiche de l'image d'origine
image = plt.imread(input_path)
 
fig, ax = plt.subplots()
im = ax.imshow(image)
ax.axis('off')
plt.show()
```

```python
# Traitement de l'image
with open(input_path, 'rb') as i:
    with open(output_path, 'wb') as o:
        input = i.read()
        output = remove(input)
        o.write(output)
```

```python
# Lecture et affiche de l'image traitée
image = plt.imread(output_path)
 
fig, ax = plt.subplots()
im = ax.imshow(image)
ax.axis('off')
plt.show()
```
